# Example plotting

This plotting example will use both casa6 and astropy.


```python
# standard plotting for python in notebooks
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# casa6 things
from casatools import image
ia = image()
```

```python
# astropy 
from astropy.visualization import simple_norm
```

```python
# example file from the M100 casaguide
fn = 'M100_Feather_CO.image.mom0.pbcor'
ia.open(fn)
data = ia.getchunk().squeeze()
ia.close()
print(data.shape)
```

```python
plt.imshow(data,origin='lower',norm=simple_norm(data, stretch='asinh'))
plt.title(fn)
plt.colorbar();
```
